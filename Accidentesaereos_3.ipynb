```python
import pandas as pd


# Ruta al archivo CSV
archivo_csv = "AccidentesAviones.csv"

# Leer el archivo CSV y almacenar los datos en un DataFrame
df = pd.read_csv(archivo_csv)

print("Columnas del DataFrame:")
print(df.columns)

# Mostrar los primeros datos del DataFrame
print("\nPrimeros datos del DataFrame:")
print(df.head())
```

```python
# Lista de columnas a eliminar
columnas_a_eliminar = [
    "HORA declarada",
    "flight_no",
    "route",
    "registration",
    "cn_ln",
    "PASAJEROS A BORDO",
    "crew_aboard",
    "ground",
    "summary",
    "Categoria",
    "Subcategoria"
]

# Eliminar las columnas especificadas del DataFrame
df = df.drop(columnas_a_eliminar, axis=1)
```

```python
# Seleccionar las columnas especificadas
df = df[['Fecha', 'Destino', 'All_aboard', 'Cantidad de Fallecidos', 'Operador', 'Ac_type', 'passenger_fatalities', 'crew_fatalities']]

# Contar el total de filas en el DataFrame
total_filas = len(df)

# Contar el total de filas que contienen el signo de interrogación (?) en cualquier columna
total_filas_con_interrogacion = len(df[df.apply(lambda row: row.astype(str).str.contains('\?', case=False).any(), axis=1)])

# Mostrar los resultados
print("Total de filas en el DataFrame:", total_filas)
print("Total de filas con el signo de interrogación (?):", total_filas_con_interrogacion)

# Mostrar las primeras 5 filas del DataFrame
print(df.head())
```

```python
import numpy as np

# Aplicar transformación logarítmica a las columnas
df['All_aboard_log'] = np.log1p(df['All_aboard'])
df['Cantidad_de_Fallecidos_log'] = np.log1p(df['Cantidad de Fallecidos'])
df['crew_fatalities_log'] = np.log1p(df['crew_fatalities'])
```

```python
# Obtener estadísticas descriptivas
estadisticas_log = df[['All_aboard_log', 'Cantidad_de_Fallecidos_log', 'crew_fatalities_log']].describe()

# Calcular cuartiles y límites para identificar outliers
q1_log = df[['All_aboard_log', 'Cantidad_de_Fallecidos_log', 'crew_fatalities_log']].quantile(0.25)
q3_log = df[['All_aboard_log', 'Cantidad_de_Fallecidos_log', 'crew_fatalities_log']].quantile(0.75)
iqr_log = q3_log - q1_log
lower_bound_log = q1_log - 1.5 * iqr_log
upper_bound_log = q3_log + 1.5 * iqr_log

# Encontrar outliers
outliers_all_aboard_log = ((df['All_aboard_log'] < lower_bound_log['All_aboard_log']) | (df['All_aboard_log'] > upper_bound_log['All_aboard_log'])).sum()
outliers_fallecidos_log = ((df['Cantidad_de_Fallecidos_log'] < lower_bound_log['Cantidad_de_Fallecidos_log']) | (df['Cantidad_de_Fallecidos_log'] > upper_bound_log['Cantidad_de_Fallecidos_log'])).sum()
outliers_crew_fatalities_log = ((df['crew_fatalities_log'] < lower_bound_log['crew_fatalities_log']) | (df['crew_fatalities_log'] > upper_bound_log['crew_fatalities_log'])).sum()

# Calcular porcentaje de outliers
total_filas_log = len(df)
porcentaje_outliers_all_aboard_log = (outliers_all_aboard_log / total_filas_log) * 100
porcentaje_outliers_fallecidos_log = (outliers_fallecidos_log / total_filas_log) * 100
porcentaje_outliers_crew_fatalities_log = (outliers_crew_fatalities_log / total_filas_log) * 100

# Imprimir resultados
print(estadisticas_log)
print(f"Porcentaje de Outliers en 'All_aboard_log': {porcentaje_outliers_all_aboard_log:.2f}%")
print(f"Porcentaje de Outliers en 'Cantidad_de_Fallecidos_log': {porcentaje_outliers_fallecidos_log:.2f}%")
print(f"Porcentaje de Outliers en 'crew_fatalities_log': {porcentaje_outliers_crew_fatalities_log:.2f}%")
```

```python
import matplotlib.pyplot as plt

# Convertir la columna 'Fecha' a objetos datetime si aún no se ha hecho
df['Fecha'] = pd.to_datetime(df['Fecha'], errors='coerce')

# Extraer el año de la columna 'Fecha'
df['Año'] = df['Fecha'].dt.year

# Contar la cantidad de accidentes por año
accidentes_por_año = df['Año'].value_counts().sort_index()

# Crear el gráfico de barras
plt.figure(figsize=(12, 6))
plt.bar(accidentes_por_año.index, accidentes_por_año.values, color='blue', alpha=0.7)
plt.title('Accidentes Aéreos por Año')
plt.xlabel('Año')
plt.ylabel('Cantidad de Accidentes')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Asegúrate de que la columna 'Fecha' esté en formato datetime
df['Fecha'] = pd.to_datetime(df['Fecha'], errors='coerce')

# Crear una nueva columna 'Año' para agrupar los datos por año
df['Año'] = df['Fecha'].dt.year

# Calcular el promedio de 'Cantidad_de_Fallecidos_log' y 'crew_fatalities_log' por año
año_promedio = df.groupby('Año').agg({
    'Cantidad_de_Fallecidos_log': 'mean',
    'crew_fatalities_log': 'mean'
}).reset_index()

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(12, 8))

# Gráfico de tendencia para 'Cantidad_de_Fallecidos_log'
ax.plot(año_promedio['Año'], año_promedio['Cantidad_de_Fallecidos_log'], label='Cantidad_de_Fallecidos_log', color='green')
# Gráfico de tendencia para 'crew_fatalities_log'
ax.plot(año_promedio['Año'], año_promedio['crew_fatalities_log'], label='crew_fatalities_log', color='red')

# Configuración de la leyenda y etiquetas de los ejes
ax.legend()
ax.set_xlabel('Año')
ax.set_ylabel('Valor Promedio (log)')
ax.set_title('Tendencias por Año de Cantidad_de_Fallecidos_log y crew_fatalities_log')

# Ajustar la disposición
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```

```python
# Agrupar por año y contar la cantidad de viajes y fallecidos por año
viajes_por_anio = df.groupby(df['Fecha'].dt.year)['All_aboard_log'].count()
fallecidos_por_anio = df.groupby(df['Fecha'].dt.year)['Cantidad_de_Fallecidos_log'].sum()

# Crear un gráfico de barras para visualizar la relación
plt.figure(figsize=(10, 6))
plt.bar(viajes_por_anio.index, viajes_por_anio.values, color='blue', alpha=0.7, label='Cantidad de Viajes')
plt.bar(fallecidos_por_anio.index, fallecidos_por_anio.values, color='red', alpha=0.7, label='Cantidad_de_Fallecidos_log')
plt.xlabel('Año')
plt.ylabel('Cantidad')
plt.title('Relación entre Cantidad de Viajes y Fallecidos por Año')
plt.legend()
plt.show()
```

```python
# Crear un gráfico de barras para la cantidad de fallecidos
plt.figure(figsize=(10, 6))
plt.bar(fallecidos_por_anio.index, fallecidos_por_anio.values, color='red', alpha=0.7, label='Cantidad de Fallecidos')
plt.xlabel('Año')
plt.ylabel('Cantidad')
plt.title('Relación entre Cantidad de Viajes y Fallecidos por Año')

# Crear un gráfico de líneas para la cantidad de viajes
plt.plot(viajes_por_anio.index, viajes_por_anio.values, color='blue', marker='o', label='Cantidad de Viajes', linewidth=2)

plt.legend()
plt.show()
```

```python
# Agrupar por año y sumar los valores
fallecidos_por_año = df.groupby(df['Fecha'].dt.year)['Cantidad_de_Fallecidos_log'].sum().reset_index()

# Calcular la correlación
correlacion_fallecidos = fallecidos_por_año['Cantidad_de_Fallecidos_log'].corr(fallecidos_por_año['Fecha'])

print(f'Correlación entre Cantidad de Fallecidos y Año: {correlacion_fallecidos}')
```
